## 3분기 누적 데이터와 당기 데이터가 같은 행에 있지 않음

```python
## 3분기 데이터와 당기 데이터를 각각 나누고 둘만 붙인 다음에 계산해야 함 - 종목에 맞게 
# 그 다음 원본 데이터에 인덱스를 기준으로 잘 붙여주기 - 당기 인덱스 기준
df['결산년도'] = [i[:4] for i in df['결산기준일']] # 붙여주기 위한 년도 열 생성
df_3Q = df[(df['당기 3분기 누적']>=1) & (df['보고서종류']=='3분기보고서')][['종목코드', '결산년도', '당기 3분기 누적']]
df_4Q = df[(df['당기']>=1) & (df['보고서종류']=='사업보고서')][['종목코드', '결산년도', '결산기준일', '당기']]
#df_3Q = df_3Q[df_3Q['당기 3분기 누적' > 0]]
df_4Q = pd.merge(df_4Q, df_3Q, on=['종목코드', '결산년도'])
df_4Q['4분기 매출'] = df_4Q['당기'] - df_4Q['당기 3분기 누적']
df = pd.merge(df, df_4Q[['종목코드', '결산기준일', '4분기 매출']], how='outer', on=['종목코드', '결산기준일'])
df
```

```python
# 다시 데이터 전처리 작업을 해줍니다. 
df.fillna('0', inplace=True) # inplace = True를 잊으면 안 됩니다!
df['4분기 매출'] = df['4분기 매출'].astype('int64')
df.info()
```

## 매출열

```python
# 매출열을 계산해줍니다.

df['매출'] = df['당기 1분기 3개월']+df['당기 반기 3개월']+df['당기 3분기 3개월']+df['4분기 매출']
```

```python
from pykrx import stock
import time
import pandas as pd
import OpenDartReader
import pandas as pd
from tqdm import tqdm
api_key = '6655d5a62b916e96636cab74f8e7e0b05692b825'
dart = OpenDartReader(api_key)
```

```python
# KOSDAQ 종목코드 조회
stock_code = stock.get_market_ticker_list(date="20210520", market="KOSDAQ")
(stock_code[:6], len(stock_code))
```

```python
def asset_f(year, corp_code, report_code):
    fs = dart.finstate_all(corp=corp_code, bsns_year=year, fs_div='OFS', reprt_code=report_code) 
    try:
        equity = int(fs.loc[fs['sj_div'].isin(['BS']) & fs['account_id'].isin(['ifrs-full_Equity']), 'thstrm_amount'].replace(",", "")) # 당기자본(자본총계)
        liability = int(fs.loc[fs['sj_div'].isin(['BS']) & fs['account_id'].isin(['ifrs-full_Liabilities']), 'thstrm_amount'].replace(",", "")) # 당기부채(부채총계)
        assets = equity + liability # 자산총계
        return assets
    except:
        equity = int(fs.loc[fs['sj_div'].isin(['BS']) & fs['account_id'].isin(['ifrs_Equity']), 'thstrm_amount'].replace(",", "")) # 당기자본(자본총계)
        liability = int(fs.loc[fs['sj_div'].isin(['BS']) & fs['account_id'].isin(['ifrs_Liabilities']), 'thstrm_amount'].replace(",", "")) # 당기부채(부채총계)
        assets = equity + liability # 자산총계
        return assets

stock_code = stock.get_market_ticker_list(date="20210520", market="KOSDAQ") # 현재일자 기준 가장 가까운 영업일의 코스닥 상장종목 리스트
res = pd.DataFrame()
start_date = '20170101'
end_date = '20201231'
for ticker in tqdm(stock_code[:100]):
    df1 = stock.get_market_ohlcv_by_date(fromdate=start_date, todate=end_date, ticker=ticker) # 기본
    df2 = stock.get_market_fundamental_by_date(fromdate=start_date, todate=end_date, ticker=ticker) # 계산값
    df = pd.concat([df1, df2], axis=1)
    df = df.assign(종목코드=ticker, 종목명=stock.get_market_ticker_name(ticker))
    res = pd.concat([res, df], axis=0)
    time.sleep(1)
res = res.reset_index()
res
```

```python
res['분기'] = pd.PeriodIndex(res['날짜'], freq='Q')
res
```

```python
res_unique = res.drop_duplicates(['종목코드', '분기'])
res_unique.reset_index(drop=True, inplace=True)  
res_unique
```

```python
quater = {'Q1': '11013',
          'Q2': '11012',
          'Q3': '11014',
          'Q4': '11011'}
res_unique['자산'] = ['' for i in range(len(res_unique['분기']))]
for i in tqdm(range(len(res_unique['자산']))):
    try:
        year = str(res_unique['분기'][i])[:4] # 년도 
        c_code = res_unique['종목코드'][i] # 회사 코드
        q_code = quater[str(res_unique['분기'][i])[-2:]] #분기 코드
        res_unique['자산'][i] = asset_f(year, c_code, q_code)
    except:
        res_unique['자산'][i] = ''
res_unique
```

```python
res_final = pd.merge(res, res_unique, how='left')
res_final.fillna(method='ffill', inplace=True)
res_final['종목분기'] = res_final['종목코드'] + res_final['분기'].astype('string') 
res_final
```

```python
df_sale['종목분기'] = df_sale['종목코드'] + df_sale['분기'].astype('string')
df_sale.info()
```

```python
df_sale.reset_index(drop=True, inplace=True)
res_final.reset_index(drop=True, inplace=True)
df = pd.merge(res_final, df_sale, how='left', on='종목분기')
df
```

```python
import warnings
warnings.filterwarnings('ignore')
import glob
import os
import datetime
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn.metrics import precision_score, recall_score, confusion_matrix, roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
from sklearn import svm

import seaborn as sns; sns.set()
```

```python
#훈련 세트와 테스트 세트를 나눈다.
#shuffle=False을 설정해 기간이 섞이지 않도록 만든다.
X_train, X_test, y_train, y_test = train_test_split(x_var, 
                                                    y_var, 
                                                    test_size=0.3, 
                                                    shuffle=False, 
                                                    random_state=3)

#훈련 세트와 테스트 세트에 포함된 양성 샘플(up) 비율을 확인한다.
train_count = y_train.count()
test_count = y_test.count()

print('train set label ratio')
print(y_train.value_counts()/train_count)
print('test set label ratio')
print(y_test.value_counts()/test_count)
```

```python
def get_confusion_matrix(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred, pos_label='positive', average='micro')
    recall = recall_score(y_test, pred,  pos_label='positive', average='micro')
    f1 = f1_score(y_test, pred,   pos_label='positive', average='micro')
    #roc_score = roc_auc_score(y_test, pred,  multi_class='ovr')
    print('confusion matrix')
    print('accuracy:{0:.4f},precision:{1:.4f},recall:{2:.4f},F1:{3:.4f}'.format(accuracy, precision, recall, f1))
```
